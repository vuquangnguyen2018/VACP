```python
import numpy as np
import pandas as pd
import openpyxl
import os 
import xlrd
# Setting Directory
os.chdir('G:/Other computers/My Laptop/VACP/')
```

```python
data_no=pd.read_excel("../VACP/Bao cao No phai thu/Báo cáo công nợ phải thu theo hạn thanh toán.xls",
                            skipfooter=1,header=6,sheet_name='Báo cáo công nợ phải thu theo h').dropna(subset=['Ngày']).reset_index(drop=True)
# data_no.loc[0:5,["Ngày","Mã","Tổng số nợ"]]
data_no.head()
```

```python
# Tạo ID Date Ngày
data_no["Năm"]=data_no["Ngày"].dt.year
data_no["Tháng"]=data_no["Ngày"].dt.month
data_no["ID Date"]=data_no["Ngày"].dt.year.astype(str)+"-"+data_no["Ngày"].dt.month.astype(str)

# Group Tổng Nợ ngày
data_no_thang=data_no.groupby(['Mã','Năm','Tháng'],as_index=False)["Tổng số nợ"].sum()
data_no_thang
```

```python
data_no_show=data_no_thang.merge(data_no[["Mã","Tên"]],on="Mã",how='left')
data_no_show=data_no_show[["Năm","Tháng","Mã","Tên","Tổng số nợ"]].drop_duplicates(subset=['Mã','Năm','Tháng'],keep='first').reset_index(drop=True)
data_no_show
```

```python
# EXPORT FILE DATA
with pd.ExcelWriter('../VACP/Bao cao No phai thu//Report.xlsx') as writer:
    data_no_show.to_excel(writer,sheet_name="Tổng hợp")
```
